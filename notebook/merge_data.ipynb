```python
import pandas as pd
import os
```

```python
dir_path = os.getcwd()
tasks_path = f"{dir_path}/tasks.csv"
userstories_path = f"{dir_path}/userstories.csv"
epics_path = f"{dir_path}/epics.csv"
tasks = pd.read_csv(tasks_path)
userstories = pd.read_csv(userstories_path)
epics = pd.read_csv(epics_path)
```

```python
# convert ref column to task_id
tasks["task_id"] = tasks["ref"]
tasks = tasks.drop(columns=["ref"])
tasks["userstory_id"] = tasks["user_story"].astype(int)
tasks = tasks.drop(columns=["user_story"])

tasks.head()
```

```python
userstories.head()
# convert ref column to userstory_id
userstories["userstory_id"] = userstories["ref"].astype(int)
userstories = userstories.drop(columns=["ref"])
userstories.head()
```

```python
epics.head()
epics["epic_id"] = epics["ref"].astype(int)
epics = epics.drop(columns=["ref"])
epics.head()
```

```python
epics["related_user_stories"]
```

```python
# extract list number after # in related_user_stories
epics["userstory_id"] = epics["related_user_stories"].str.findall(r"#(\d+)")
epics = epics.drop(columns=["related_user_stories"])

epics = epics.explode("userstory_id")
epics["userstory_id"] = epics["userstory_id"].astype(int)
epics["epic_name"] = epics["subject"]
epics = epics.drop(columns=["subject"])
epics.head()
```

```python
merge_df = pd.merge(
    tasks,
    userstories,
    on="userstory_id",
    how="left",
    suffixes=(
        "_task",
        "_userstory",
    ),
    validate="m:1",
)
```

```python
merge_df.head()
```

```python
tasks.shape, userstories.shape, merge_df.shape
```

```python
merge_df = pd.merge(
    merge_df,
    epics,
    on="userstory_id",
    how="left",
    suffixes=("", "_epic"),
    validate="m:1",
)
```

```python
merge_df.shape
```

```python
# save to excel with sheet name merge_data
with pd.ExcelWriter("merge_data.xlsx") as writer:
    merge_df.to_excel(writer, sheet_name="merge_data", index=False)
```
